```python
r = requests.get("https://www.timeanddate.com/weather/usa/north-dakota")
```

```python
soup = bs(r.text, 'html.parser')
```

```python
soup
```

```python
tags = soup.find_all('a', href=True)
```

```python
num_locations = soup.findAll(attrs={"class": "small soft"})
txt = num_locations[0].getText()
num = int(re.search(r'\d+', txt).group())
```

```python
tags
```

```python
city_hrefs = [h for h in tags if "/weather/usa/" in h["href"]]
```

```python
lst = city_hrefs[-num:]
lst[0]['href']
```
